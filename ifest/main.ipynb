```python
import numpy as np
import pandas as pd
```

```python
f1 = pd.read_excel("fitur_1.xlsx",sheet_name="Table format")
f2 = pd.read_excel("fitur_2.xlsx",sheet_name="Table format")
f3 = pd.read_excel("fitur_3.xlsx",sheet_name="Table format")
target = pd.read_excel("target.xlsx",sheet_name="Table format")
```

```python
f1f2 = pd.merge(f1,f2,on="GeoAreaName",how="inner")
```

```python
f3t = pd.merge(f3,target, on="GeoAreaName", how="inner")
```

```python
merged = pd.merge(f1f2,f3t, on="GeoAreaName", how="inner")
```
